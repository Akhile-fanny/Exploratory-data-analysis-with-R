```r
#loading the packages and dataset we will be using for this data exploration 
library(tidyverse) 
library(ggplot2) 
library(gapminder)
head(gapminder) # show sthe first 6 rows in the dataset
```

```r
# summary statistics of the data
gapminder %>%
summarize(avg_pop= mean(pop),
          max_pop = max(pop), 
          max_lifeExp= max(lifeExp), 
          min_lifeExp = min(lifeExp),  
          avg_lifeExp = mean(lifeExp))
```

```r
#highest life expentancy by country in 2007, selecting all except the continent
gapminder %>%
filter (year == 2007)%>%
arrange(desc(lifeExp))%>% 
group_by(country)%>%
select( - continent)

# NOTE THAT JAPAN HAS THE HIGHEST LIFE EXPECTANCY IN 2007 WITH THE VALUE OF 82.603
```

```r
#continent with the lowest life exp in 1987

gapminder %>%
filter(year == 1987)%>%
group_by(continent) %>%
arrange(lifeExp)


# NOTE THAT ANGOLA HAS THE LOWEST LIFE EXPECTANCY IN 1987 WITH THE VALUE OF 39.906

```

```r

# population distribution in the countries
ggplot(gapminder, aes(x = year, y = pop, color = continent)) + 
geom_point() + 
scale_x_log10() + 
facet_wrap(~ continent) + 
ggtitle( "population distribution in the countries")

```

```r
#data for only 2007
lifeexp_2007 <-
gapminder %>%
filter(year == 2007)
```

```r
# how does the population change by gdp per capital ?
ggplot(lifeexp_2007, aes(x=pop , y =gdpPercap , color = continent)) + 
geom_point() + 
scale_x_log10() + ggtitle("population by gdp per capital")
```

```r
# how does the population affects life expectancy  ?
ggplot(lifeexp_2007, aes(x=pop , y =lifeExp , color = continent)) + 
geom_point() + 
scale_x_log10() + ggtitle("population by life expectancy")
```
